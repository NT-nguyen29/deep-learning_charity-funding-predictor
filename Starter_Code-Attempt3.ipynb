```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("Resources/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(columns=["EIN", "NAME"])
application_df.head()
```

```python
# Determine the number of unique values in each column
# (https://stackoverflow.com/questions/56310788/different-outcome-using-pandas-nunique-and-unique)
application_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
app_count = application_df["APPLICATION_TYPE"].value_counts()
app_count
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_types_to_replace = list(app_count[app_count < 500].index)
application_types_to_replace
```

```python
# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
class_count = application_df["CLASSIFICATION"].value_counts()
class_count
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
class_count_1 = class_count[class_count > 1]
class_count_1
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_to_replace = list(class_count[class_count < 100].index)
classifications_to_replace
```

```python
# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
numeric_df = pd.get_dummies(application_df)
numeric_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
# target: IS_SUCCESSFUL, features: everything else
y = numeric_df["IS_SUCCESSFUL"]
X = numeric_df.drop(["IS_SUCCESSFUL"], axis=1)

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
number_input_features = len(X_train_scaled[0])
number_input_features
```

```python
# (borrowed from 21.2, activity 4)
# Create a method that creates a new Sequential model with hyperparameter options
def create_model(hp):
    nn_model = tf.keras.models.Sequential()

    # Allow kerastuner to decide which activation function to use in hidden layers
    activation = hp.Choice('activation',['relu','tanh','sigmoid'])
    
    # Allow kerastuner to decide number of neurons in first layer
    nn_model.add(tf.keras.layers.Dense(units=hp.Int('first_units',
        min_value=1,
        max_value=30,
        step=5), activation=activation, input_dim=number_input_features))

    # Allow kerastuner to decide number of hidden layers and neurons in hidden layers
    for i in range(hp.Int('num_layers', 1, 6)):
        nn_model.add(tf.keras.layers.Dense(units=hp.Int('units_' + str(i),
            min_value=1,
            max_value=30,
            step=5),
            activation=activation))
    
    nn_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

    # Compile the model
    nn_model.compile(loss="binary_crossentropy", optimizer='adam', metrics=["accuracy"])
    
    return nn_model
```

```python
# Import the kerastuner library (for optimizing model)
import kerastuner as kt

tuner = kt.Hyperband(
    create_model,
    objective="val_accuracy",
    max_epochs=20,
    hyperband_iterations=2)
```

```python
# Run the kerastuner search for best hyperparameters
tuner.search(X_train_scaled,y_train,epochs=20,validation_data=(X_test_scaled,y_test))
```

```python
optimal_hps = tuner.get_best_hyperparameters(num_trials=1)[0]
optimal_hps
```

```python
# retrieve optimal hyperparameters
# (https://www.tensorflow.org/tutorials/keras/keras_tuner)
optimal_hps = tuner.get_best_hyperparameters(num_trials=1)[0]

print(f"Optimal number of units in first layer: {optimal_hps.get('first_units')}")
print(f"Number of layers: {optimal_hps.get('num_layers')}")
print(f"Optimal activation: {optimal_hps.get('activation')}")
```

```python
# find optimal number of epochs by building model with optimal hyperparameters we found
model = tuner.hypermodel.build(optimal_hps)
history = model.fit(X_train_scaled, y_train, epochs=50, validation_data=(X_test_scaled,y_test))

val_acc_per_epoch = history.history['val_accuracy']
optimal_epoch = val_acc_per_epoch.index(max(val_acc_per_epoch)) + 1
print("Optimal number of epochs: %d" % (optimal_epoch))
```

```python
# reinstantiate the model, this time using the optimal hyperparameters and epochs
model2 = tuner.hypermodel.build(optimal_hps)
model2.fit(X_train_scaled, y_train, epochs=optimal_epoch, validation_data=(X_test_scaled,y_test))
```

```python
# evaluate the model
eval_result = model2.evaluate(X_test_scaled, y_test)
print("[test loss, test accuracy]:", eval_result)
```

```python
# Export our model to HDF5 file
model2.save("Models/attempt_3/AlphabetSoupCharity.h5")
```
